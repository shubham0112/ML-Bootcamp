```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

import cv2
```

```python
frame = cv2.imread("thor.jpg")
```

## cv2 reads in BGR

```python
plt.imshow(frame)
```

```python
# so we convert it back to rgb
rgb = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb)
```

## Lets talk about faces!!

```python
# haarcascade is the data of faces, so we pass that into the cascade classifier
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
# it gives x,y,width,height of the faces we provide
# since here only one face(of thor) so only one coordinates will be there
faces = classifier.detectMultiScale(rgb)
```

```python
face = faces[0]
```

```python
x, y, w, h = face
```

```python
face = rgb[y:y+h, x:x+w]
```

```python
plt.imshow(face)
```

## Now lets draw the rectangle on the image itself

```python
out = cv2.rectangle(rgb,(x,y),(x+w,y+h),(0,0,255),2)
```

```python
plt.imshow(out)
```
