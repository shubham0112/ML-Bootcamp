## linspace and plotting

```python
x = np.linspace(0,2*np.pi,100)
```

```python
y = np.sin(x)
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
```

```python
x = [4, 7, 2, 5]
y = [3, 2, 5, 1]
plt.plot(x,y)
```

## circle

```python
r = 10
```

```python
theta = np.linspace(0,2*np.pi,200)
```

```python
x = r*np.cos(theta)
y = r*np.sin(theta)
```

```python
plt.plot(x,y)
# we are getting oval bcoz we are stretching the axes
```

```python
plt.figure(figsize=(4,4))
plt.plot(x,y)
```

## scatter plot

```python
x = np.linspace(2,20,1000)
```

```python
m,c = 3,10
```

```python
y = m*x + c
```

```python
plt.scatter(x,y)
```

```python
noise = np.random.randn(1000)*5
```

```python
y_modified = y+noise
```

```python
plt.scatter(x,y_modified)
```

## histogram

```python
x = np.random.randn(1000)
# randn gives normal dist. with mean=0 and sd=1(approx)
```

```python
out = plt.hist(x,bins=120)
# max values are 0 bcoz the mean is zero
# notice that most of the values is between -1 to 1 bcoz the sd is 1
```

## pie chart

```python
out = plt.pie([10,20,30],explode=[0,.1,0])
```

## Seaborn lib (use it along with matplotlib to get best results)

```python
import seaborn as sns
```

```python
x = np.random.randn(1000)
```

```python
plt.figure(figsize=(4,4))
sns.displot(x,bins=30,kde=True,rug=True)
```

```python
data = sns.load_dataset("tips")
```

```python
data.head()
```

```python
sns.scatterplot(x=data["total_bill"],y=data["tip"],hue=data["smoker"])
```

```python
sns.violinplot(y=data.total_bill)
# so we can see that most of the values is between 10 and 20
```

## scikit learn

```python
from sklearn.datasets import load_boston
```

```python
X,y = load_boston(return_X_y=True)
```

```python
sns.violinplot(y=y)
# here most of the values is between 15-25
# value of 50 can be considered as outlier( i.e something very differebt from the normal data)
```
